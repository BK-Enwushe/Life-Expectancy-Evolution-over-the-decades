## HOW LIFE EXPECTANCY HAS EVOLVED ACROSS CONTINENTS OVER THE DECADES 

```python
import numpy as np
import pandas as pd
import seaborn as sb
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
# Loading dataset
df = pd.read_csv("C:/Users/BUQUEKELLS/Documents/data/gapminder-FiveYearData.csv")
df
```

```python
"""
# Creating pivot table containing year (x-axis) and continent (y-axis)
the average life expectancy across the various continents was computed against the years. 
Where ‘continents’ represents the row and ‘year’ the column. The ‘life expectancy’ in the 
syntax represents the values.
"""
pivot_table = df.pivot_table(index="continent", columns="year", values="lifeExp")
pivot_table
```

```python
# Creating HeatMap -- ‘annot=True’ is the average life expectancy values distributed within the heatmaps AND ‘cmap=None’ represents the colour palletes which was not uttered.
# Heat_Map = sb.heatmap(pivot_table, annot=True, cmap=None)
Heat_Map = sb.heatmap(pivot_table, annot=True, cmap="Reds")
Heat_Map.set_title("Life Expectancy by Continent and Year")
plt.show()
```
